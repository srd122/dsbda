```python
import numpy as np 
import pandas as pd 
import seaborn as sb 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('Heart.csv')
df
```

```python
air = pd.read_csv('AirQuality.csv',sep=";")
air
```

ERROR CORRECTING

```python
df.isna().sum()
```

```python
df1 = df.fillna(method='bfill')
df1
```

```python
air.isnull().sum()
```

```python
air1 =air.fillna(method='ffill')
air1
```

DATA TRANSFORMATION

```python
df1.loc[df1['Sex']==1,'Sex']='M'
df1.loc[df1['Sex']==0,'Sex']='F'
df1
```

```python
air1['Date'] = pd.to_datetime(air1['Date'])

#To Replace the Comma's with Dot
air1.replace(to_replace=',',value='.', regex=True, inplace=True)
air1
```

DATA MODEL BUILDING

```python
plt.figure(figsize=(10,10))
sb.heatmap(data = df1.corr(),cbar=True,square=True,annot=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
attribute = df[['Age','Chol','MaxHR']]
target = df['RestECG']
train_x,test_x,train_y,test_y=train_test_split(attribute,target,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,r2_score,mean_absolute_error
lr = LogisticRegression()
lr.fit(train_x,train_y)
lr_y_pred = lr.predict(test_x)
lr_acc = accuracy_score(test_y,lr_y_pred)
lr_acc
```
